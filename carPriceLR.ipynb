```python
# label encoder convert text to numbers 
from sklearn.preprocessing import LabelEncoder
lbl = LabelEncoder()
```

```python
data.fueltype = lbl.fit_transform(data.fueltype)
data.aspiration = lbl.fit_transform(data.aspiration)
data.doornumber = lbl.fit_transform(data.doornumber)
data.carbody = lbl.fit_transform(data.carbody)
data.drivewheel = lbl.fit_transform(data.drivewheel)
data.enginelocation = lbl.fit_transform(data.enginelocation)
data.enginetype = lbl.fit_transform(data.enginetype)
data.cylindernumber = lbl.fit_transform(data.cylindernumber)
data.fuelsystem = lbl.fit_transform(data.fuelsystem)
```

```python
'''
RobustScaler in scikit-learn
can help minimize the effect of outliers during model training without explicitly removing them
'''
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
data_scaled = scaler.fit_transform(data[['price']])
```

```python
x = data.iloc[:, :22].values
y = data.iloc[: ,-1].values
```

```python
x
```

```python
y.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
xtrain , xtest , ytrain ,ytest = train_test_split(x, y , test_size =.3 , random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor 
regressor = RandomForestRegressor(random_state=0)
regressor.fit(xtrain,ytrain)
ypred = regressor.predict(xtest)
```

```python
data = pd.DataFrame(data={'Predicted Profit' : ypred.flatten().round(2)})
```
